```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv("Data_Iris.csv")
data.head()
```

```python
# Assigning the numerical features here
features = data.iloc[:, 0:4].values

# Using the last 3 features for plotting
X_plot = features[:, 1:4]
```

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(features)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_plot[:, 0], X_plot[:, 1], X_plot[:, 2], c=kmeans_labels, cmap='viridis', s=50)
ax.set_xlabel('Sepal Width (cm)')
ax.set_ylabel('Petal Length (cm)')
ax.set_zlabel('Petal Width (cm)')
ax.set_title('K-Means Clustering (3D)')
plt.show()
```

```python
# Agglomerative Hierarchical Clustering
agg = AgglomerativeClustering(n_clusters=3)
agg_labels = agg.fit_predict(features)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_plot[:, 0], X_plot[:, 1], X_plot[:, 2], c=agg_labels, cmap='plasma', s=50)
ax.set_xlabel('Sepal Width (cm)')
ax.set_ylabel('Petal Length (cm)')
ax.set_zlabel('Petal Width (cm)')
ax.set_title('Agglomerative Hierarchical Clustering (3D)')
plt.show()
```
