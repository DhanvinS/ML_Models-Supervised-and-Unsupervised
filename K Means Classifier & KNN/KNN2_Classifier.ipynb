```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
#import statements
```

```python
df = pd.read_csv('Data_Iris.csv')
#setting the data from iris dataset
df.head()
```

```python
df.columns = df.columns.str.strip()
#removing any leading or trailing whitespace from all column names in our DataFrame
```

```python
sb.pairplot(df, vars=df.columns[0:4], hue='species_name', diag_kind='hist')
plt.show()
#plotting pairplot to visualize our data
```

```python
#split into independent variables and target/dependent variable

x = df.iloc[:,:4]     #independent variables
y = df["species_name"]  #target variables
```

```python
x_train, x_test, y_train , y_test = train_test_split(x, y, train_size = 120, test_size = 30, stratify=y)
#splitting data according to whats given in the question
```

```python
# Function to train and evaluate KNN for different k values
def KNN(k):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train, y_train)
  y_predicted = knn.predict(x_test)

  #Testing Accuracy
  accuracy = accuracy_score(y_test, y_predicted)
  print(f"Test Accuracy: {accuracy:.4f}")

  #Getting Consusion Matrix
  cm = confusion_matrix(y_test, y_predicted, labels = knn.classes_)
  print(pd.DataFrame(cm, index = knn.classes_, columns = knn.classes_))
```

```python
#checking for k values of 3 , 5 and 7

for k in [3, 5, 7]:
  KNN(k)
```
