```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df.head()
```

```python
inputs = df.drop("salary_more_then_100k", axis = 'columns')
target = df["salary_more_then_100k"]
```

Need to divide dataframe into target and independent variable
inputs = independent variables
target = target variable or dependent variables

We add the axis parameter to the drop method in pandas to specify whether we want to remove a row or a column.

If we set axis='columns' or axis=1, pandas will remove a column.

If we set axis='index' or axis=0, pandas will remove a row.

```python
from sklearn.preprocessing import LabelEncoder as LE
```

Convert strings into numbers using label encoders as machine cant make decision based on strings

```python
le_company = LE()
le_job = LE()
le_degree = LE()
# creating obejcts for 3 different columns
```

```python
inputs['company_le'] = le_company.fit_transform(inputs['company'])
inputs['job_le'] = le_job.fit_transform(inputs['job'])
inputs['degree_le'] = le_degree.fit_transform(inputs['degree'])
```

This method does two things:  

 1 It learns all unique company names the 'fit',  
 2 It converts each company name into a number the 'transform'. For example, 'google' might become 0, 'abc' might become 1, and so on.

```python
inputs.head()
```

```python
inputs_le = inputs.drop(['company', 'job', 'degree', 'company_n', 'job_n', 'degree_n'], axis = 'columns')
inputs_le.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_le, target)
```

```python
model.score(inputs_le, target)
```

```python
model.predict([[2, 2, 1]])
```
