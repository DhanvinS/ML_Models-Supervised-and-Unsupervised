```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
plt.xlabel('Ages')
plt.ylabel('Probability of taking insurance(1, 0)')
plt.scatter(df.age, df.bought_insurance , marker="+", color = "green" )
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.age
y = df.bought_insurance
```

```python
x_train , x_test, y_train, y_test = train_test_split(x, y, train_size = 0.9, test_size = 0.1)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train.values.reshape(-1, 1), y_train)

'''
x_train = x_train.values.reshape(-1, 1) is doing two things:

.values: Converts your pandas Series or DataFrame to a NumPy array.

         Makes it compatible with scikit-learn’s fit methods.

.reshape(-1, 1): Changes the shape of the array to have one column and as many rows as needed.

         -1 automatically infers the number of rows from the data.

          1 sets the number of columns (features) to one.

Result:
If x_train has 100 samples, it becomes a 2D array with shape (100, 1).

'''
```

```python
df.tail()
```

```python
lr.predict(x_test.values.reshape(-1, 1))
```

```python
lr.score(x_test.values.reshape(-1, 1), y_test)
```
