```python
import pandas as pd


df = pd.read_csv("Data_Iris.csv")
df.head()
```

```python
# Separate features and labels
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y = df['species_name']
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

# Function to run t-SNE and plot
def plot_tsne(X, y, perplexity):
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    X_embedded = tsne.fit_transform(X)

    # Create a DataFrame for plotting
    df_embedded = pd.DataFrame(X_embedded, columns=['Dim1', 'Dim2'])
    df_embedded['species'] = y

    # Plot
    plt.figure(figsize=(6,5))
    sns.scatterplot(x='Dim1', y='Dim2', hue='species', data=df_embedded, palette='Set1', s=60)
    plt.title(f"t-SNE with Perplexity={perplexity}")
    plt.show()

# Run for different perplexities
for p in [5, 10, 25]:
    plot_tsne(X, y, perplexity=p)
```

Here are my comments on the Effect of Perplexity

Refering my plots


Perplexity = 5:
t-SNE focuses on very local structure (small neighborhoods). Clusters might look tighter, but global relationships may be distorted. Some points may appear more scattered.

Perplexity = 10:
Balances local and global structure better. Clusters begin to separate more clearly while preserving neighborhood relationships.

Perplexity = 25:
Larger neighborhood is considered. Clusters are smoother and better separated globally, but very fine local distinctions may be lost.

Brief:

Low perplexity  captures local relationships, can show small clusters but may fragment the global structure.

Medium perplexity  balances local and global structure.

High perplexity  captures global structure, clusters may look more rounded and well-separated.
