```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
from google.colab import files
files.upload()
```

```python
df = pd.read_csv('Data_Iris.csv')
df.head()
```

```python
x = df.iloc[ : , :-1]
y = df.iloc[ : , -1]
#assigning all columns except last column to x and the last column to y which is the target variable
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 120, test_size = 30)
```

```python
# for 2(a) question
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver = 'lbfgs')
lr.fit(x_train, y_train)
```

```python
y_predicted_a = lr.predict(x_test)
```

```python
test_accuracy_a = accuracy_score(y_test, y_predicted_a)
test_accuracy_a
```

```python
cm = confusion_matrix(y_test, y_predicted_a)
plt.figure(figsize = (6, 4))
class_names = sorted(y.unique())  # get class names from my y variable
sb.heatmap(cm, cmap='Blues',
           annot=True,
           xticklabels=class_names,
           yticklabels=class_names)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion matrix for 2(a)')
```

```python
# for 2(b) question
from sklearn.svm import LinearSVC
svc = LinearSVC(C=10000000000, max_iter=2000)
svc.fit(x_train, y_train)
 #As C increases, the model initially improves until it fits noise,
 #after which test accuracy may drop because of overfitting.
 #in this case increasing c does not seem to affect test accuracy
```

```python
y_predicted_b = svc.predict(x_test)
```

```python
test_accuracy_b = accuracy_score(y_test, y_predicted_b)
test_accuracy_b
```

```python
cm2 = confusion_matrix(y_test, y_predicted_b)
plt.figure(figsize = (6, 4))
class_names = sorted(y.unique())  # get class names from my y variable
sb.heatmap(cm2, cmap='Greens',
           annot=True,
           xticklabels=class_names,
           yticklabels=class_names)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion matrix for 2(b)')
```

```python
# for 2(c) question
from sklearn.svm import SVC
rb_svc = SVC(kernel='rbf', C=1.0)
rb_svc.fit(x_train, y_train)
```

```python
y_predicted_c = rb_svc.predict(x_test)
```

```python
test_accuracy_c = accuracy_score(y_test, y_predicted_c)
test_accuracy_c
```

```python
cm3 = confusion_matrix(y_test, y_predicted_c)
plt.figure(figsize = (6, 4))
class_names = sorted(y.unique())  # get class names from my y variable
sb.heatmap(cm3, cmap='Reds',
           annot=True,
           xticklabels=class_names,
           yticklabels=class_names)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion matrix for 2(c)')
```
